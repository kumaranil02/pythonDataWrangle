```python
def create_data(PolicyNumber,dos,doe):
    global data_mart
    doe1 =doe
    dos1 = dos
    check = 0

    while dos < doe1:
        data_store = pd.DataFrame()
        if dos.year == 2015 and check == 0:
            temp = dos
            dos = datetime.datetime(2015,1,1)
            doe = temp
            check = 1
        elif dos.day == 1 and dos.month == 1:
            doe = dos1.replace(day = dos1.day,month = dos1.month,year=dos.year+1)
        else:
            doe = dos.replace(day=1,month=1,year=dos.year+1)
        if dos.year == doe1.year and dos.year != 2015:
            doe = doe1
        data_store['PolicyNumber'] = [PolicyNumber]
        data_store['COV PHASE'] = 1
        data_store['CLNDR YR'] = dos.year
        data_store['DRTN STR DT'] = dos
        data_store['DRTN END DT'] = doe
        data_mart = data_mart.append(data_store)
        dos = doe
        
    return "Complete"
```

```python
data_mart = pd.DataFrame()

data['DOI'] = data['Date of Issue'].mask(data['Date of Issue'].dt.year < 2015,
                                           data['Date of Issue'] + pd.offsets.DateOffset(year=2015))

data['Status'] = data.apply(lambda x : create_data(x['PolicyNumber'],x['DOI'],x['Status Date']),axis=1)
```
